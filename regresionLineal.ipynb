Qu algoritmo vamos a usar en vez de regresin logstica? As es! Regresin lineal.
La idea general de la regresin lineal es que hay una sencilla ecuacin lineal que describe cmo un determinado parmetro define el output. Primero, imaginemos que la imagen de abajo nos muestra cmo se relaciona el precio con el rea:

y es el objetivo, una variable dependiente que queremos encontrar. En este caso, es el precio. x es la caracterstica, una variable independiente que define a la dependiente. Aqu es el rea. Fjate que, a diferencia de la regresin logstica, el lado derecho de la ecuacin afecta directamente a la variable dependiente en vez de mediante una curva log-odds. Por eso es una lnea recta.

a representa la pendiente y muestra dos cosas. En primer lugar, si es positiva el precio aumenta junto con el rea; si es negativa, el precio baja conforme el rea aumenta. En segundo lugar, muestra cunto cambia el precio si el rea aumenta por cada unidad. En el grfico de arriba, el precio aumenta aproximadamente 5 unidades por cada 30 unidades del rea. Eso significa que la pendiente es 5/30 o 1/6.

Lleva las caractersticas de entrenamiento a la variable features y la caracterstica last_price del objetivo a la variable target. Divide el valor de la caracterstica target entre 1000000.

Inicializa un modelo de regresin lineal y entrnalo. Calcula el valor de la mtrica RECM en el conjunto de validacin y almacnalo en la variable result.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

df = pd.read_csv('datasets/moved_train_data_us.csv')

# Extraer caractersticas y objetivos
features = df.drop(['last_price'], axis=1)  # Todas las columnas excepto 'last_price'
target = df['last_price'] / 1000000  # Normalizar la columna objetivo

features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.25, random_state=12345) # segmenta el 25% de los datos para hacer el conjunto de validacin

model = LinearRegression() # inicializa el constructor de modelos
model.fit(features_train, target_train) # entrena el modelo en el conjunto de entrenamiento
predictions_valid = model.predict(features_valid) # obtn las predicciones del modelo en el conjunto de validacin

result = mean_squared_error(target_valid, predictions_valid) ** 0.5# calcula la RECM en el conjunto de validacin
print("RECM del modelo de regresin lineal en el conjunto de validacin:", result)
```
